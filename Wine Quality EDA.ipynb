```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
wine = pd.read_csv('winequality.csv', sep = ';')
wine.head()
```

```python
wine.info()
```

```python
wine.describe()
```

```python
plt.figure(figsize=(20,10))
c=wine.corr()
sns.heatmap(c,cmap="BrBG",annot=True)
```

```python
#sns.set(style="darkgrid")

sns.countplot(x='quality', data=wine)

# Quality range is between 3 and 8.
```

```python
The quality of wine ranges from 3 to 8. 
We can observe that most of the wine quality is in the mid range which is 5 and 6.
```

```python
conditions = [
    (wine['quality'] < 5),
    (wine['quality'] >= 7)
]
rating = ['low','high']
wine['rating'] = np.select(conditions, rating,default='normal' )
wine['rating'].value_counts()
```

```python
wine['rating'].value_counts().plot(kind='bar',color='Blue')
plt.title("Quality Rating")
plt.ylabel('count')
plt.xlabel('quality')
```

```python
We have divided the wine quality into three types.

Quality  = less than 5 , Low

Quality = 5 and 6,  Normal

Quality = 7 and Above,  High
```

```python
fig = plt.figure(figsize = (10,6))
sns.barplot(x = 'quality', y = 'density', data = wine)
```

```python
Density does  not affect the quality of the wine.
```

```python
fig = plt.figure(figsize = (10,6))
sns.barplot(x = 'quality', y = 'residual sugar', data = wine)
```

```python
Residual sugar does not affect the quality of wine in a significant way.
```

```python
fig = plt.figure(figsize = (10,6))
sns.barplot(x = 'quality', y = 'citric acid', data = wine)
```

```python
Good quality wines have a higher level of citric acid.
```

```python

fig = plt.figure(figsize = (10,6))
sns.barplot(x = 'quality', y = 'alcohol', data = wine)
```

```python
Good quality wines have higher amount of alcohol.
```

```python

fig = plt.figure(figsize = (10,6))
sns.barplot(x = 'quality', y = 'chlorides', data = wine)
```

```python
Good quality wines have less amount of chlorides.
```

```python
fig = plt.figure(figsize = (10,6))
sns.barplot(x = 'quality', y = 'volatile acidity', data = wine)
```

```python
Good quality wines have lower volatile acidity.
```

```python
fig = plt.figure(figsize = (10,6))
sns.barplot(x = 'quality', y = 'sulphates', data = wine)
```

```python
Good quality wines have higher sulphates content but the difference is not very significant.
```

```python
fig = plt.figure(figsize = (10,6))
sns.barplot(x = 'quality', y = 'volatile acidity', data = wine)
# volatile acidity decrease with quality
```

```python
sns.lmplot(x = "alcohol", y = "residual sugar", col = "rating", data = wine)
```

There is a linear relationship between alcohol and residual sugar for low quality wines. This means that the residual sugar value increase along with alcohol content in low quality wines. In high and medium quality wines residual sugar does not change with alcohol content. This can be a a very crucial factor to avoid low quality wines.

```python
bx = sns.boxplot(x="rating", y='citric acid', data = wine)
bx.set(xlabel='Wine Ratings', ylabel='Citric Acid', title='Citric acid in different types of Wine ratings')
```

```python
Citric acid content is high in good quality wines and less in low quality wines.
```

```python
bx = sns.boxplot(x="rating", y="pH", data = wine);
bx.set(xlabel='Wine Ratings', ylabel='pH', title='pH in different types of Wine ratings')
```

```python
The pH value is high in low quality wines.
pH is  almost same in both normal and high quality wines.
```

```python
bx = sns.boxplot(x="rating", y='alcohol', data = wine)
bx.set(xlabel='Wine Quality', ylabel='Alcohol Percent', title='Alcohol percent in different wine quality types')
```

Alcohol content is greater in high quality wines.
It is almost similar in low and normal quality wines.

Exploratory data analysis helped us in determining the factors which affect the quality of the wine both in a positive and negative manner. We can use these insights to predict the quality of wine when we build our machine learning model.
